```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("election_result_2024-2.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

Key Insights from dataset:
We are gathering key insights from the dataset, such as the total number of constituencies in the Lok Sabha election, details of seats won by the BJP, INC, and other parties, the top 10 parties in terms of winning seats, and ## the distribution of vote share percentage. These insights help us understand the overall election results and the performance of major political parties.

```python
total_constituencies = df['constituency_name'].nunique()
print("Total constituencies in Lok Sabha election 2024:", total_constituencies)
```

Details of seats won by Bhartiya Janata Party (BJP)

```python
df.loc[(df['party_name'] == 'Bharatiya Janata Party') & (df['rank'] == 1)]
```

```python
df.loc[(df['party_name'] == 'Indian National Congress') & (df['rank'] == 1)]
```

```python
df.loc[(df['party_name'] != 'Indian National Congress') & (df['party_name'] != 'Bharatiya Janata Party')& (df['rank'] == 1)]
```

```python
winners = df[df['rank'] == 1]
seats_won = winners['party_name'].value_counts()
top_10_parties = seats_won.head(10)
seats_won_final = top_10_parties

plt.figure(figsize=(12, 8))
ax = seats_won_final.plot(kind='barh', color='blue')
plt.title('Number of Seats Won by Each Party (Top 10)')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.show()
```

```python
total_votes_by_party = df.groupby('party_name')['total_votes'].sum()
top_10_votes = total_votes_by_party.nlargest(10)


plt.figure(figsize=(12, 8))
top_10_votes.plot(kind='pie', autopct='%1.1f%%', title='Votes % Distribution by Party')
plt.ylabel('')
plt.show()
```

```python
# Group by state and count number of seats won
state_seats = winners['state_name'].value_counts()

# Top 10 states with most seats
top_10_states = state_seats.head(10)
print("Top 10 states with most seats:\n", top_10_states)

# Bottom 10 states with least seats
bottom_10_states = state_seats.tail(10)
print("\nBottom 10 states with least seats:\n", bottom_10_states)
```

```python
# Plotting Top 10
plt.figure(figsize=(12, 8))
top_10_states.plot(kind='barh', color='green')
plt.title('Top 10 States with Most Lok Sabha Seats')
plt.xlabel('Number of Seats')
plt.ylabel('State')
plt.show()
```

```python
# Plotting Bottom 10
plt.figure(figsize=(12, 8))
bottom_10_states.plot(kind='barh', color='red')
plt.title('Bottom 10 States with Least Lok Sabha Seats')
plt.xlabel('Number of Seats')
plt.ylabel('State')
plt.show()
```

```python
# Define the Seven Sister states
seven_sisters = [
    "Arunachal Pradesh",
    "Assam",
    "Manipur",
    "Meghalaya",
    "Mizoram",
    "Nagaland",
    "Tripura"
]
```

```python
# Filter winners from only these states
seven_sisters_winners = winners[winners['state_name'].isin(seven_sisters)]

# Count seats per state
seven_sisters_seats = seven_sisters_winners['state_name'].value_counts()
```

```python
print("Seats won in Seven Sister States:\n")
print(seven_sisters_seats)

# Plot graph
plt.figure(figsize=(10, 6))
seven_sisters_seats.plot(kind='bar', color='purple')
plt.title("Lok Sabha Seats Won in Seven Sister States (2024)")
plt.xlabel("State")
plt.ylabel("Number of Seats")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count seats per state/UT
state_seats = winners['state_name'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(14, 12))
state_seats.sort_values(ascending=True).plot(kind='barh', color='teal')

plt.title("Lok Sabha Seats Distribution by States and Union Territories (2024)")
plt.xlabel("Number of Seats")
plt.ylabel("State / Union Territory")
plt.tight_layout()
plt.show()
```

```python
party_state = winners.groupby(['state_name','party_name']).size().unstack(fill_value=0)

party_state.plot(
    kind='barh', 
    stacked=True, 
    figsize=(14,12), 
    colormap='tab20'
)

plt.title("Party-wise Seat Distribution by State (2024)")
plt.xlabel("Number of Seats")
plt.ylabel("State/UT")
plt.show()
```
